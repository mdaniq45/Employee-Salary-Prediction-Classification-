### Step 3: Data Preprocessing

```python
#handilling  missing values
df['Salary'].fillna(df['Salary'].mean(),inplace=True)
```

### Step-by-Step Pipeline Creation for Multiple Algorithms


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier, GradientBoostingRegressor, RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, accuracy_score
from sklearn.neighbors import KNeighborsRegressor
import xgboost as xgb
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
```

```python
# Split the data into features (X) and target (y)
X = df_encoded.drop(columns=['Salary', 'Is_Active', 'ID', 'Joining_Date'])
y_reg = df_encoded['Salary']  # For regression
y_class = df_encoded['Is_Active']  # For classification

# Train-test split for regression and classification tasks
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X, y_reg, test_size=0.2, random_state=42)
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(X, y_class, test_size=0.2, random_state=42)
```

```python
# Numeric features preprocessing pipeline (imputation + scaling)
numeric_features = X.select_dtypes(include=['int64', 'float64']).columns
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

# Categorical features preprocessing pipeline (imputation + encoding)
categorical_features = X.select_dtypes(include=['object']).columns
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Full preprocessor combining numeric and categorical
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Regression pipeline with model choices
regression_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', RandomForestRegressor(random_state=42))  # Default model, change as needed
])

# Train and evaluate multiple regression models
models = {
    'Decision Tree': DecisionTreeRegressor(),
    'Random Forest': RandomForestRegressor(),
    'KNN': KNeighborsRegressor(),
    'SVR': SVR(),
    'Gradient Boosting': GradientBoostingRegressor(),
    'XGBoost': xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100)
}

for model_name, model in models.items():
    # Update the model in the pipeline
    regression_pipeline.set_params(model=model)
    
    # Train the model
    regression_pipeline.fit(X_train_reg, y_train_reg)
    
    # Make predictions
    y_pred_reg = regression_pipeline.predict(X_test_reg)
    
    # Evaluate the model
    mse = mean_squared_error(y_test_reg, y_pred_reg)
    print(f"{model_name} Regression Mean Squared Error: {mse}")
```

### Step 4: Build a Pipeline for Classification Models


```python
# Classification pipeline with models
classification_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', RandomForestClassifier(random_state=42))  # Default model, change as needed
])

# Train and evaluate multiple classification models
classification_models = {
    'Random Forest Classifier': RandomForestClassifier(),
    'Naive Bayes': GaussianNB(),
    'Logistic Regression': LogisticRegression(),
    'SVM Classifier': LogisticRegression(),
}

for model_name, model in classification_models.items():
    # Update the model in the pipeline
    classification_pipeline.set_params(model=model)
    
    # Train the model
    classification_pipeline.fit(X_train_class, y_train_class)
    
    # Make predictions
    y_pred_class = classification_pipeline.predict(X_test_class)
    
    # Evaluate the model
    accuracy = accuracy_score(y_test_class, y_pred_class)
    print(f"{model_name} Classification Accuracy: {accuracy}")
```
